```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
from random import randint
```

```python
img = plt.imread("base.jpg")
FINAL_SIZE = (1500, 1500)
PATCH_SIZE = (150, 250)
print(img.shape)
```

```python
final_img = np.zeros((FINAL_SIZE[0], FINAL_SIZE[1], 3), dtype=np.uint8)

def get_random_patch(img, PATCH_SIZE):
    i = randint(0, img.shape[0] - PATCH_SIZE[0])
    j = randint(0, img.shape[1] - PATCH_SIZE[1])
#     res = img[i:i + PATCH_SIZE[0], j:j + PATCH_SIZE[1]]
    return img[i : i + PATCH_SIZE[0], j : j + PATCH_SIZE[1]]


for i in range(0, FINAL_SIZE[0], PATCH_SIZE[0]):
    for j in range(0, FINAL_SIZE[1], PATCH_SIZE[1]):
        final_img[i : i + PATCH_SIZE[0], j : j + PATCH_SIZE[1]] = get_random_patch(img, PATCH_SIZE)

plt.imshow(final_img)
plt.imsave("im1.jpg", final_img)
```

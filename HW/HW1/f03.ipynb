```python
cv2.destroyAllWindows()
cv2.namedWindow('first', cv2.WINDOW_NORMAL,)
cv2.namedWindow('second', cv2.WINDOW_NORMAL,)
cv2.namedWindow('third', cv2.WINDOW_NORMAL,)
cv2.namedWindow('base', cv2.WINDOW_NORMAL,)

img = np.array(cv2.imread("C:/Users/nimac/MyProjects/vision/HW/HW1/BW.tif"))
row, col = img.shape[0:2]
img1 = np.uint32(img[:row//3, :, :])
img2 = np.uint32(img[row//3: 2*row//3, :, :])
img3 = np.uint32(img[2*row//3:row, :, :])
row = row//3
# cv2.imshow("first", img1)
# cv2.imshow("second", img2)
# cv2.imshow("third", img3)
# cv2.waitKey(0)
```

```python
# start = (row//4, col//2)
# end = (row//2, 3*col//4)
start = (1000, 1200)
end = (1500, 1600)
def calc_MSE(mat1, mat2):
    return np.sum(np.square(mat1 - mat2))

def cut(mat, start, end):
    return mat[start[0]:end[0], start[1]:end[1]]


def find_best_pos(img1, img2, start, end, row, col, row_s, row_e, col_s, col_e):
    min_err2 = 1000000000
    min_err2_row, min_err2_col = (0, 0)
    for i in tqdm(range(row_s, row_e)):
        for j in range(col_s, col_e):
            mat1 = cut(img1[:, :, 0], start, end)
            mat2 = cut(img2[:, :, 0], (start[0] + i, start[1] + j), (end[0] + i, end[1] + j))
            err = calc_MSE(mat1, mat2)
            if err < min_err2:
                min_err2 = err
                min_err2_row = i
                min_err2_col = j
    return min_err2_row, min_err2_col

row2, col2 = find_best_pos(img1, img2, start, end, row, col, -50, 0, 0, 20)
row3, col3 = find_best_pos(img1, img3, start, end, row, col, -150, -100, -5, 15)

```
